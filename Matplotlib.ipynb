```python
from matplotlib import pyplot as plt
print(plt.style.available)
plt.style.use('ggplot')
ages_x = list(range(25,60,5))
dev_y = list(range(30000,100000,10000))
py_dev =[45372, 53850, 57287, 63016, 70003, 71496, 83640]
plt.plot(ages_x, dev_y, color='b', marker='.', label='Everyone')
plt.plot(ages_x,py_dev, color='y', marker='<',label='Py')
plt.title('Median Salary (USD) by Age')
plt.xlabel('Ages')
plt.ylabel('Median Salary')
plt.legend()
plt.tight_layout()
#plt.show()
```

```python
plt.style.use('fivethirtyeight')
ages_x = list(range(25,60,5))
dev_y = list(range(30000,100000,10000))
py_dev =[45372, 53850, 57287, 63016, 70003, 71496, 83640]
plt.plot(ages_x, dev_y, label='Everyone')
plt.plot(ages_x,py_dev, label='Py')
plt.title('Median Salary (USD) by Age')
plt.xlabel('Ages')
plt.ylabel('Median Salary')
plt.legend()
plt.tight_layout()
plt.grid(False)
plt.show()
```

```python

import numpy as np
plt.style.use('ggplot')
ages_x = list(range(25,60,5))
indices = np.arange(len(ages_x))
width = 0.25
dev_y = list(range(30000,100000,10000))
py_dev =[45372, 53850, 57287, 63016, 70003, 71496, 83640]
plt.bar(indices, dev_y, width=width, label='Everyone')
plt.bar(indices+width,py_dev, width=width, label='Py')
#plt.title('Median Salary (USD) by Age')
#plt.xlabel('Ages')
#plt.ylabel('Median Salary')
plt.legend()
plt.xticks(ticks=indices, labels=ages_x)
plt.tight_layout()
plt.grid(False)
#plt.show()
plt.savefig('play.png')
```
