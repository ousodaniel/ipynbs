## 2.1.11 Sampleing and Inference Foundation
Sample need to be representative of the population
Sample (statistic) distribution
- average of multiple sample distribution means is equal to $\mu$ (population mean)
- std (${s}$) deviation of multiple sample distribution means is equal to $\frac{\sigma}{\sqrt{n}}$ (a.k.a standard error of $\bar {X}$)
- irrespective of the population distribution the above two conditions hold

Regardless of the population distribution, the sampling distribution of the sample mean will approach normal distribution as the sample size gets bigger.

### 2.1.14: Estimation
Observations from the sample can be carried over to the population: statistic to parameter
- point estimation of the population parameter
- interval estimation; lies within a range and with what confidence

Begin from a point estimate toward the interval estimate
The confidence of a population inervl estimate is often taken as 95% (it is possible to see other levels). It means that for a repeated sampling we can be 95% sure that the statistic's interval estimate will contain the parameter.
A higher confidence level means a wider interval.

```python
samp = np.array([120, 55, 60, 10, 8, 150, 44, 58, 62, 123])
x_bar = np.mean(samp)
s_samp = np.std(samp)

sim_data = np.linspace(samp.min(), samp.max(), 50)
pdf = stats.norm.pdf(sim_data, x_bar, s_samp)
sns.displot(pdf, kind='kde')
```

```python
# sample statistic
x_bar, s_std = 110, 7

# sample size
n = 50

# construct the CI
error = round(s_std / np.sqrt(n), 2)
ci = np.round(stats.norm.interval(alpha=0.95, loc=x_bar, scale=error), 2)
ci
```

```python
Inn the above, we have assumed a normal distribution of the population and used sample standard deviation (s).
However, often, th epopulation sigma is unkown (although can be estimated from the sample) and the population distribution is mostly not normal.
To adjust for the preceeding facts, we adjust to a t-distribution, whee we assume the sample sigma, for the population sigma: and correct for that.

t vs. normal distribution:
    - wider variance
    - lower peak
    - as the degrees of freedom increases, the t-dist tend towards a normal distrubution
```

```python
x = np.linspace(-3, 3, 100)

fig, axes = plt.subplots(2, 3, sharey=True, figsize=(15 ,10))
axes = axes.ravel()

for i,k in zip(range(6), [1, 2, 3, 5, 10, 30]):
    ax = axes[i]
    ax.plot(x, stats.t.pdf(x, df=k), color='b', label='t-dist')
    ax.plot(x, stats.norm.pdf(x), color='r', label='norm-dist')
    ax.set_title(f't-distribution for {k} degrees of freedom')
    ax.legend()
plt.tight_layout()
```

#### Hypothesis formulation

- Null: nothing acting on the status quo (asociated with =, >=, <=) Newton's firt law - stationary body
- Alt: An external force acting significantly (asociated with !=, >, <) Newton"s second law - impact = magnitude & direction
- The space of both hypotheses got to be be the same

Test statistic
- Used for the Decision Rule
- normal, t, F, Chi-square

Hypothesis testing errors
- **Type I error ($Prob=\alpha$)**: Rejecting H<sub>0</sub> when H<sub>0</sub> is True
- **Correct decision ($Prob=1-\beta$)**: Rejecting H<sub>0</sub> when H<sub>0</sub> is False
- **Correct decision ($Prob=1-\alpha$)**: Fail to reject H<sub>0</sub> when H<sub>0</sub> is True
- **Type II error ($Prob=\beta$)**: Fail to reject H<sub>0</sub> when H<sub>0</sub> is False


### 2.2.4 Template for Hypothesis Testing

1. Identify the key question
2. Establish the hypothesis
3. Understand and prepare data
4. Identify the right test statistic
5. Check the assumptions: does the data satisfy the assumptions of the test statistic
6. Perform the test

### 2.2.5 Performing a Hypothesis Test
- Level of significance ($\alpha$): how wrong are we permitting ourselves to be wrong. Set before hand (usually 5%)
- p-value: how likely it the test statitic if the H<sub>0</sub> is the case. How much support do we have for the H<sub>0</sub> hypothesis
- Acceptance/Rejection Region: respectively partition the area under the distribution curve of the test statistic and determine where the calculated test statistic lies

```python
# population parameters
n, mu, sigma = 45, 5, 1.3

# sample mean
x_bar = 5.25

# calc test stat
std_error = sigma / np.sqrt(n)
test_stat = (x_bar - mu) / std_error
test_stat
```

## Regression and Prediction

[**Formulation**](https://en.wikipedia.org/wiki/Linear_regression)

Given a data set ${\{x_{i1},...,x_{ip}, y_i\}^n_{i=1}}$ of $n$ observations, a linear model takes the form

$y_i=\beta_0+\beta_1x_{i1}+...+\beta_px_{ip}+\epsilon_i$

Which can be written in a vector form as

$y_i=x_i^T\beta+\epsilon_i$,

where $x_i^T$ is a column vector of the $i^{th}$ observation of all the explanatory variables; $\boldsymbol {\beta }$ is a $p\times 1$ vector of unknown parameters; and the scalar $\epsilon_{i}$ represents unobserved random variables (errors) of the $i^{th}$ observation.

Which can be further written in a matrix notation as $n$ stacked equations

${\displaystyle \mathbf {y} =\mathrm {X} {\boldsymbol {\beta }}+{\boldsymbol {\varepsilon }},\,}$

where

$\mathbf {y} ={\begin{pmatrix}y_{1}\\y_{2}\\\vdots \\y_{n}\end{pmatrix}},\quad $

${\displaystyle X={\begin{pmatrix}\mathbf {x} _{1}^{\mathsf {T}}\\\mathbf {x} _{2}^{\mathsf {T}}\\\vdots \\\mathbf {x} _{n}^{\mathsf {T}}\end{pmatrix}}={\begin{pmatrix}1&x_{11}&\cdots &x_{1p}\\1&x_{21}&\cdots &x_{2p}\\\vdots &\vdots &\ddots &\vdots \\1&x_{n1}&\cdots &x_{np}\end{pmatrix}},}\ \boldsymbol {X}\ $ is a matrix of $n$-dimensional column-vectors $X_{j}$ or of row-vectors ${x} _{i},$

${\displaystyle {\boldsymbol {\beta }}={\begin{pmatrix}\beta _{0}\\\beta _{1}\\\beta _{2}\\\vdots \\\beta _{p}\end{pmatrix}},\quad {\boldsymbol {\varepsilon }}={\begin{pmatrix}\varepsilon _{1}\\\varepsilon _{2}\\\vdots \\\varepsilon _{n}\end{pmatrix}}.}$

**Markov Chain**

Used to compute probabilities of events occurring, by viewing them as transitioning states. A probabilistic automaton includes the probability of a given transition into the transition function, turning it into a transition matrix.

- Reducibility: A Markov chain is irreducible if it is possible to get to any state from any state, or if there is a chain of steps between any two states that have a positive probability.

- Periodicity: A state in a Markov chain is periodic if the chain can return to the state only at multiples of some integer larger than 1. Thus, starting in state 'i', the chain can return to 'i' only at multiples of the period 'k', and k is the largest such integer. State 'i' is aperiodic if k = 1 and periodic if k > 1.

- Transience and Recurrence: A state 'i' is said to be transient if, given that we start in state 'i', there is a non-zero probability that we will never return to 'i'. State i is recurrent (or persistent) if it is not transient. A recurrent state is known as positive recurrent if it is expected to return within a finite number of steps and null recurrent otherwise.

- Ergodicity: A state 'i' is said to be ergodic if it is aperiodic and positive recurrent. If all states in an irreducible Markov chain are ergodic, then the chain is said to be ergodic.

- Absorbing State: A state ‘i' is called absorbing if it is impossible to leave this state. Therefore, the state 'i' is absorbing if pii = 1 and pij = 0 for i ≠ j. If every state can reach an absorbing state, then the Markov chain is an absorbing chain.

***Central Limit Theorem***

Normalised error: the estimated score minus the true score divided by the square root of the number of observations.

$$\frac{\hat{x}-x}{\sqrt n}$$

By Gaussian, the variation in the estimate (the error) varies as:

$Error=TrueScore(1-TrueScore)$

If the true score is one or zero, the error is going to be zero; if it is close to half, then the error is going to be large. Since we do not know apriori, we assume the error to be the largest; in the case of a coin toss 0.5. Therefore, the true score is most likely to be in the range

$$EstScore+\frac{0.5}{\sqrt n}\gt TrueScore \gt EstScore-\frac{0.5}{\sqrt n}$$

Therefore the correction term would generaly be

$$\frac {Constant (C)}{\sqrt {n (NumObs)}}$$

But how do we choose $C$? For a coin toss the trues score ranges form zero to one, thus average score is 0.5. Similarly, for rating scores ranging from "zero" to five 2.5 is a great $C$.

The adjusted score in the bias scenario above would be (conservative):

$$AverageRating-\frac {Constant (C)}{\sqrt {n (NumObs)}}$$

However, if the ratings are only few but non-biased the adjusted score would be (optimistic):

$$AverageRating+\frac {Constant (C)}{\sqrt {n (NumObs)}}$$

Another correction type also need consideration; the freshness of ratings/reviews. How do we account for it?

We normalise the observation (say time in case of a restaurant review) then the weight would be: $w=exp(-NormalisedAge)$. The normalised age is negated to signify that as age increases, weight decreases.

The weighted average score of reviews (R) is:

$$R_{weighted}=\frac{\sum_{i=1}^n W_iS_i}{\sum_{i=1}^n W_i}$$

Consequently, the correction in this average score would be:

$$\frac{C}{\sqrt{\frac{\sum_{i=1}^n W_i}{W_{max}}}}$$

### Influence maximisation in/of a network
Will approach it from a Active-Inactive (corresponding to Suceptible-Infected)

Influence of a subset of nodes is taken as the expected number of nodes that eventually becomes active (at stabilisation).

How many nodes finally become active for agood model for the contagion process can be achievd inexpensively by a Monte-Carlo approach; simulate the process spreading through the network a repeated number of times and find the average of the number of nodes that ultimately become active.

- number of runs for each node ($n$) is logarithmic to $n$
- assuming $n$ nodes and $m$ edges
- each edge contributes, there a simulation of the contagion process requires at most $m$ computations
- therefore the total computation cost would be in the order $m\ast n\ast log(n)$

While the above is for a single node, the process is would be similar for a subset of nodes $k$. However, this would be impossible to search over. Therefore, we increnmentaly build a subset of influencial nodes in a greedy manner (node after node) exploiting the MOnte-Carlo approach.

If, say, we need a set $K$ with $k$ nodes the iteration will stop after $k$ iterations. Therefore the computatinal cost would be $k(CostOneNode)=k\ast m\ast n\ast log(n)$

The influence of the resultant $K$ is guranteed to be almost two-thirds of the best possible subset.

#### Cross-Validation
To measure performance of a ML model
Principle: partition the data into fixed number of partitions/folds; apply the model to each and average the error estimate across the folds

**Types of Cross-Validation**
Broadly:
- Exhaustive
- Non-exhaustive

**Non -exhaustive Methods**

Holdout: train-test split
- Usually the training set is more than twice the test set
- Commonly the ratio 7:3 or 8:2 are used
- Best for a very large datset, quick baseline builds

K-fold: randomly partition data into $k$ equal folds; train on $k-1$ folds while holdimg out the $k^{th}$ for testing; test on the $k^{th}$ fold; repeat the "train-test" cycle until all $k$s are used as test set; average the resulting $k$ accuracies to make your cross-validation acuracy
- Generally results in lesser bias, as every datapoint participates in training
- Among the best when data is small
- Cons: $k$ computations from scratch; may be problematic with classification problems

Stratified K-fold: first rearranges the data so that each fold wil be a good representation of the whole; perform K-fold

**Exhaustive Methods**

Leave-P-Out: takes out $p$ (part) datpoints out of $n$ datapoints in th e dataset; train model on $n-p$ data points and test on $p$; iterate for all possible combinations of $p$; average across iterations

$${n\choose k} = {\frac{n!}{k!(n-k)!}}$$

Leave-One-Out: simplifies Leave-P-Out so that $p=1$, thus $n$ number of iterations; less exhaustive

Rolling: for time series; creates folds in a forward chaining manner
- Due to randomness in corss-validation partitioning the order of the time searies data, and therefore its crux, will be interfered with
- The result of the above would mean our training set will have data point belonging to the future, which is what we want to predict
